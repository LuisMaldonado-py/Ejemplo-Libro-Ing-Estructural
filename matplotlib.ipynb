```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("beam.csv")
x=np.array(data.x)
M1=np.array(data.M1)
M2=np.array(data.M2)
V1=np.array(data.V1)
V2=np.array(data.V2)
```

```python
plt.plot(x,M1)
plt.show()
```

```python
fig, ax=plt.subplots()
ax.plot(x, M1)
plt.show()
```

```python
plt.plot(x,M1, label="M1")
plt.xlabel("x")
plt.ylabel("bending moment")
plt.title("Beam")
plt.legend()
plt.gca().invert_yaxis()
plt.gcf().set_size_inches(6,3)
plt.show()
```

```python
fig, ax=plt.subplots()
ax.plot(x,M1, label="M1")
ax.set_xlabel("x")
ax.set_ylabel("bending moment")
ax.set_title("Beam")
ax.legend()
ax.invert_yaxis()
fig.set_size_inches(6,3)
plt.show()
```

```python
fig, ax=plt.subplots(figsize=(8,3))
ax.invert_yaxis()
ax.plot(x,M1, label="M1", color="deeppink", linestyle="--", linewidth=3)
plt.show()
```

```python
fig, ax=plt.subplots(figsize=(8,3))
ax.invert_yaxis()
ax.plot(x,M1, label="M1", linewidth=3)
ax.plot(x,M2, label="M2", linewidth=3)
plt.show()
```

```python
plt.style.use("seaborn")
fig, ax=plt.subplots(figsize=(8,3))
ax.invert_yaxis()
ax.plot(x,M1, label="M1")
ax.plot(x,M2, label="M2")
plt.show()
```

```python
plt.style.use("seaborn-whitegrid")
fig, ax=plt.subplots(figsize=(8,3))
ax.invert_yaxis()
ax.plot(x,M1, label="M1", linewidth=3)
ax.plot(x,M2, label="M2", linewidth=3)
ax.fill_between(x, M1, alpha=0.3)
ax.fill_between(x, M2, alpha=0.3)
plt.show()
```

```python
fig, ax=plt.subplots(figsize=(8,3))
ax.invert_yaxis()
ax.plot(x,M1, label="M1", linewidth=3)
ax.plot(x,M2, label="M2", linewidth=3)
ax.fill_between(x, M1,M2, alpha=0.3, interpolate=True)
plt.show()
```

```python
fig, ax=plt.subplots(figsize=(8,3))
ax.invert_yaxis()
ax.plot(x,M1, label="M1", linewidth=3)
ax.fill_between(x, 0, M1, alpha=0.3, where=(M1<=0))
ax.fill_between(x, 0, M1, alpha=0.3, where=(M1>=0))
plt.show()
```

```python
fig, ax=plt.subplots(figsize=(8,3))
ax.invert_yaxis()
ax.plot(x,M1, label="M1")
ax.plot(x,M2, label="M2")
ax.fill_between(x, 0, M1, alpha=0.3)
ax.fill_between(x, 0, M2, alpha=0.3)
ax.plot([0,4,9],[0,0,0], linewidth=2, color="dimgrey", 
        marker=6, markersize=16)
plt.show()
```

```python
fig, (ax1, ax2)=plt.subplots(nrows=2, ncols=1, sharex=True, figsize=(8, 4))
ax1.plot(x,M1, label="M1")
ax1.set_ylabel("bending moment")
ax1.set_title("Combo 1")
ax1.invert_yaxis()

ax2.plot(x,M2, label="M2")
ax2.set_ylabel("bending moment")
ax2.set_title("Combo 2")
ax2.invert_yaxis()
plt.show()
```

```python
fig, ((ax1, ax2),(ax3,ax4))=plt.subplots(nrows=2, ncols=2, sharex=True)
fig.set_size_inches(8, 4)
ax1.plot(x,M1, label="M1")
ax1.set_ylabel("bending moment")
ax1.set_title("Combo 1")
ax1.invert_yaxis()

ax2.plot(x,V1, label="V1")
ax2.set_title("Combo 1")

ax3.plot(x,M2, label="M2")
ax3.set_ylabel("bending moment")
ax3.set_title("Combo 2")
ax3.invert_yaxis()

ax4.plot(x,V2, label="V2")
ax4.set_title("Combo 2")
plt.show()
```

```python
raindata=pd.read_csv("raindata.csv")
province=np.array(raindata["province"])
T=np.array(raindata["T"])
daily=np.array(raindata["daily prec."])
ndays=np.array(raindata["days exceeding 10 mm 2007-2016"])
```

```python
plt.style.use("seaborn-whitegrid")
fig, ax=plt.subplots()
ax.scatter(daily,ndays, c=T, cmap="Blues", marker="p")
ax.set_xlabel("daily prec.")
ax.set_ylabel("days exceeding 10 mm")
plt.show()
```

```python
fig, ax=plt.subplots()
im=ax.scatter(daily,ndays, c=T, cmap="Blues", marker="p")
ax.set_xlabel("daily prec.")
ax.set_ylabel("days exceeding 10 mm")
cbar=fig.colorbar(im, ax=ax)
cbar.set_label("avg. temperature")
plt.show()
```

```python
plt.scatter(daily,ndays, c=T, cmap="Blues", marker="p")
plt.xscale("log")
cbar=plt.colorbar()
plt.show()
```

```python
fig, ax = plt.subplots()
ax.bar(province[:15], daily[:15])
plt.draw()
ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
ax.set_ylabel("Avg. temperature")
plt.show()
```

```python
barWidth=0.3
r1=np.arange(len(province[:15]))
r2=r1+barWidth

fig, ax = plt.subplots()
ax.bar(r1, daily[:15], width=barWidth, label="daily avg.")
ax.bar(r2, ndays[:15], width=barWidth, label="days > 10mm")
ax.set_xticks(r1)
ax.set_xticklabels(province[:15], rotation=45)
ax.set_ylabel("Avg. temperature")
ax.legend()
plt.show()
```
